```python
# Examining usage of conditionals
import numpy as np
data = np.loadtxt(fname='./data/inflammation-01.csv', delimiter=',')

max_inflammation_0 = np.amax(data, axis=0)[0]
max_inflammation_20 = np.amax(data, axis=0)[20]

if max_inflammation_0 == 0 and max_inflammation_20 == 20:
    print('Suspicious looking maxima!')
elif numpy.sum(numpy.amin(data, axis=0)) == 0:
    print('Minima add up to zero!')
else:
    print('Seems OK!')
```

```python
# Special conditional cases
if '':
    # empty string is FALSE
    print('empty string is true')
if 'word':
    # nonempty string is TRUE
    print('word is true')
if []:
    # empty list is FALSE
    print('empty list is true')
if [1, 2, 3]:
    # nonempty list is TRUE
    print('non-empty list is true')
if 0:
    # zero is FALSE
    print('zero is true')
if 1:
    # nonzero is TRUE
    print('one is true')
```

```python
# write conditional to test if value is within 10% of another

close_value = 9
far_value = 5
target = 10

acceptable_bounds = [target - (target * 0.1), target + (target * 0.1)]
print(acceptable_bounds)

if acceptable_bounds[0] <= close_value <= acceptable_bounds[1]:
    print("Close Value Accepted") 

if acceptable_bounds[0] <= far_value <= acceptable_bounds[1]:
    print("Far Value Accepted") 
```

```python
# In place operators
x = 1  # original value
x += 1 # add one to x, assigning result back to x
x *= 3 # multiply x by 3
print(x)

# sum multiple lists at the same time using in place operators
pos_list = list(range(1,100,2))
neg_list = list(range(-100,0,2))
sum = 0
for i in range(max(len(pos_list), len(neg_list))):
    if i < len(pos_list):
        sum += pos_list[i]
    if i < len(neg_list):
        sum += neg_list[i]
    
# expecting -50 (largest pos is 99, smallest negative is -100 (-1 sum), 
#                counts up/down by twos, 50 pairs across list)
print(sum)
```

```python
# Sorting filenames
filenames = ['inflammation-01.csv',
            'myscript.py',
            'inflammation-02.csv',
            'small-01.csv',
            'small-02.csv']
large_files = []
small_files = []
other_files = []

for fname in filenames:
    if fname.lower().startswith('infl'):
        large_files.append(fname)
    elif fname.lower().startswith('small'):
        small_files.append(fname)
    else:
        other_files.append(fname)

print(large_files)
print(small_files)
print(other_files)
```

```python
# count vowels in passed string
vowels = ['a', 'e', 'i', 'o', 'u']
words = ['erythrocytosis', 'laryngeal', 'polymerase', 'inflammation']
num_vowels = 0
for word in words:
    no_vowel = True
    contains_y = False
    for letter in word:
        if letter in vowels:
            num_vowels += 1
            no_vowel = False
        if letter == 'y':
            contains_y = True
    if no_vowel and contains_y:
        num_vowels += 1

print(num_vowels)
```
