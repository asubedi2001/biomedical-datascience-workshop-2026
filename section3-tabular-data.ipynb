```python
import os
import numpy as np
import pandas as pd
```

```python
fp_inflammation = './data/inflammation-'

# row -> patient, col -> day
df_data1 = pd.read_csv(fp_inflammation + '01.csv', header=None)
```

```python
data = df_data1.to_numpy()
print(data)
```

```python
import matplotlib.pyplot
image = matplotlib.pyplot.imshow(data)
matplotlib.pyplot.show()
```

```python
ave_inflammation = np.mean(data, axis=0)
ave_plot = matplotlib.pyplot.plot(ave_inflammation)
matplotlib.pyplot.show()
```

```python
max_plot = matplotlib.pyplot.plot(np.amax(data, axis=0))
matplotlib.pyplot.show()
```

```python
min_plot = matplotlib.pyplot.plot(np.amin(data, axis=0))
matplotlib.pyplot.show()
```

```python
import matplotlib.pyplot as plt

data = np.loadtxt(fname='./data/inflammation-01.csv', delimiter=',')

fig = plt.figure(figsize=(10.0, 3.0))

axes1 = fig.add_subplot(1, 3, 1)
axes1.set_title("Average Inflammation")
axes2 = fig.add_subplot(1, 3, 2)
axes2.set_title("Max Inflammation")
axes3 = fig.add_subplot(1, 3, 3)
axes3.set_title("Min Inflammation")


axes1.set_ylabel('average')
axes1.plot(np.mean(data, axis=0))

axes2.set_ylabel('max')
axes2.plot(np.amax(data, axis=0))

axes3.set_ylabel('min')
axes3.plot(np.amin(data, axis=0))

fig.tight_layout()

plt.savefig('inflammation.png')
plt.show()
```
