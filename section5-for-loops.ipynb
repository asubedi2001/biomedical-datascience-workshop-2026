```python
# Summation over a list in Python
# Calculate the sum of the numbers in a list
numbers = [124, 402, 36]
summed = 0
for number in numbers:
    summed += number

print(summed)

import numpy as np
summed = np.sum(numbers)
print(summed)
```

```python
# Utilising enumerate
# Example: Evaluation of polynomials

x = 5
coefs = [2, 4, 3]
y = coefs[0] * x**0 + coefs[1] * x**1 + coefs[2] * x**2
print(y)

result = 0
for idx, val in enumerate(coefs):
    result += val * pow(x, idx)
 
print(result)
```
