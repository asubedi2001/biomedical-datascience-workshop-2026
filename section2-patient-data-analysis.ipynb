```python
import os
import numpy as np
import pandas as pd
```

```python
fp_inflammation = './data/inflammation-'

# row -> patient, col -> day
df_data1 = pd.read_csv(fp_inflammation + '01.csv', header=None)
```

```python
p1_ndarray = df_data1.to_numpy()
print(p1_ndarray)
```

```python
print(type(p1_ndarray[0][0]))
print(type(p1_ndarray[0, 0]))
```

```python
print(p1_ndarray[0:4, 0:10])
```

```python
small = p1_ndarray[:3, 36:]
print('small is:')
print(small)
```

```python
print(np.mean(p1_ndarray))

maxval, minval, stdval = np.amax(p1_ndarray), np.amin(p1_ndarray), np.std(p1_ndarray)
print('maximum inflammation:', maxval)
print('minimum inflammation:', minval)
print('standard deviation:', stdval)
```

```python
patient_1 = p1_ndarray[0, :] # 0 on the first axis (rows), everything on the second (columns)
print('maximum inflammation for patient 0:', np.amax(patient_1))
```

```python
print('maximum inflammation for patients:\n', np.amax(p1_ndarray, axis=1))
print('maximum daily inflammation aross patients:\n', np.amax(p1_ndarray, axis=0))
```

```python
print(np.mean(p1_ndarray, axis=0))
print(np.mean(p1_ndarray, axis=0).shape)
print(np.mean(p1_ndarray, axis=1))
```
